```python
#For 2017
df17 = pd.read_csv('PhoraDurbarKathmandu_PM2.5_2017_YTD.csv')
```

```python
df17
```

```python
# df17['NowCast Conc.'] = df17['NowCast Conc.'].replace({-999.0:''})

df17['NowCast Conc.'] = df17['NowCast Conc.'].replace([-999.0], np.nan)
```

```python
column = {'NowCast Conc.' : 'conc'}
```

```python
df17 = df17.rename(columns = column)
```

```python
df17
```

```python
df17 = df17.dropna(subset=['conc'])
```

```python
df17
```

```python
#Setting colors
def build_colors(row):
    if row['conc'] < 50:
        return 'green'
    elif row['conc'] < 100:
        return 'yellow'
    elif row['conc'] < 150:
        return 'orange'
    elif row['conc'] < 200:
        return 'red'
    elif row['conc'] < 300:
        return 'brown'
    elif row['conc'] < 500:
        return 'darkmagenta'
    else:
        return 'darkmagenta'

colors17 = df17.reset_index().apply(build_colors, axis=1)
```

```python
df17.plot.scatter(x='Hour', y='conc', color=colors17)

plt.savefig("2017_byhour.pdf")
```

```python
df17.plot.scatter(x='Day', y='conc', color = colors17)

plt.savefig("2017_by_day.pdf")
```

```python
df17.plot.scatter(x='Month', y='conc',color = colors17)

plt.savefig("2017_by_month.pdf")
```
